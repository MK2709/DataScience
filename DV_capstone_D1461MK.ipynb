# #Determines



#### Features

- timestamp - timestamp field for grouping the data
- cnt - the count of a new bike shares
- t1 - real temperature in C
- t2 - temperature in C “feels like”
- hum - humidity in percentage
- wind_speed - wind speed in km/h
- weather_code - category of the weather
- is_holiday - boolean field - 1 holiday / 0 non holiday
- is_weekend - boolean field - 1 if the day is weekend
- season - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.

**"weather_code" category description:**
* 1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity 
* 2 = scattered clouds / few clouds 
* 3 = Broken clouds 
* 4 = Cloudy 
* 7 = Rain/ light Rain shower/ Light rain 
* 10 = rain with thunderstorm 
* 26 = snowfall 
* 94 = Freezing Fog

### 6.Make feature engineering. Extract new columns (day of the week, day of the month, hour, month, season, year etc.) 
###feature engineering ile anlatilmak istenilen sey featurelarimiz yani sutunlarimizdan yeni turev sutunlar olusturma. yani timestamp sutunundan baska sutunlar turetecegiz.

you can use strftime('%Y-%m') for year_mount

```python
df.index
```

```python
df.index.year #indexteki sadece yillari cekti
```

```python
f= lambda x: x.strftime("%Y-%m") #burada lamda fonksiyonu ile f degiskenine tanimladik
df['year_month']=f(df.index) #f fonksiyonunu index sutunundaki butun degerlere uyguladik ve bunu yeni bir stuna yazdirdik
df['year']=df.index.year #indexteki sadece yillari cekti
df['month']=df.index.month #indexten sadece aylari cekti
df['day_of_month']=df.index.day #indexten sadece gunleri cekti
df['day_of_week']=df.index.dayofweek #indexten sadece dayofweekleri cekti
df['hour']=df.index.hour #indexten sadece hourlari cekti
```
