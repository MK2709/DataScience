```python
#Funtional Method
x=np.linspace(0,10,10)
y1=x
y2=x**2
y3=x**3
y4=np.sqrt(x)

#Functional Method
plt.figure()
plt.subplot(2,2,1) #alt grafikleri 2x2 matrix halinde olacak ve bu 1 nolu grafik
plt.plot(x,y1,'r-.')#cizilecek grafigin x,y degerleri olacagi kirmizi(r) ve dashdot oalcagi
plt.title('$y_1=x$') #cizdigimiz grafigin basligi 
plt.xlabel('x')
plt.ylabel('$y_1$')


plt.subplot(2,2,2) #alt grafikleri 2x2 matrix halinde olacak ve bu 2 nolu grafik
plt.plot(x,y2,'b-*')#cizilecek grafigin x,y degerleri olacagi kirmizi(r) ve dashdot oalcagi
plt.title('$y_2=x^2$') #cizdigimiz grafigin basligi 
plt.xlabel('x')
plt.ylabel('$y_2$')

plt.subplot(2,2,3) #alt grafikleri 2x2 matrix halinde olacak ve bu 32 nolu grafik
plt.plot(x,y3,'g-+')#cizilecek grafigin x,y degerleri olacagi kirmizi(r) ve dashdot oalcagi
plt.title('$y_3=x^3$') #cizdigimiz grafigin basligi 
plt.xlabel('x')
plt.ylabel('$y_3$')

plt.subplot(2,2,4) #alt grafikleri 2x2 matrix halinde olacak ve bu 32 nolu grafik
plt.plot(x,y2,'b-x')#cizilecek grafigin x,y degerleri olacagi kirmizi(r) ve dashdot oalcagi
plt.title('$y_4=\sqrt{x}$') #cizdigimiz grafigin basligi KAREKOK isaretine DIKKAT
plt.xlabel('x')
plt.ylabel('$y_4$')

plt.tight_layout() #goruntunun duzenli olmasini sagliyor.
#herzaman kullan
```

```python
x=np.linspace(0,10,10)
y1=x
y2=x**2
y3=x**3
y4=np.sqrt(x)
```

```python
from matplotlib import axes
fig,axes = plt.subplots(2,1)
axes[0].plot(x,y1)
axes[0].set_title("$X=Y_1$")
axes[0].set_xlabel("x")
axes[0].set_ylabel("$y_1$")
axes[1].plot(x,y2)
axes[1].set_title("$X=Y_2$")
axes[1].set_xlabel("Y")
axes[1].set_ylabel("$Y_2$")
plt.tight_layout()
```

```python
fig,axes = plt.subplots(1,4)
axes[0].plot(x,y1,label='$x=y_1$')
axes[0].set_title("$X=Y_1$")
axes[1].plot(x,y2)
axes[2].plot(x,y3)
axes[3].plot(x,y4)
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style


style.use('grayscale') #kullanima hazir styl paketi
plt.grid(True)
x=np.arange(0,30,0.2)
y=np.sin(x)
plt.plot(x,y)
```

```python
style.use('default')
x=np.arange(0,30,0.2)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1, label='Sin(x)')
plt.plot(x,y2, label='Cos(x)')
plt.legend(loc='upper right')

plt.title("$sin(x)$ Function")
plt.suptitle("I'm the biggest title")
plt.xlabel("$X$ Values")
plt.ylabel("$y$ Values")
```

```python
blood_sugar_man = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women=[67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```
